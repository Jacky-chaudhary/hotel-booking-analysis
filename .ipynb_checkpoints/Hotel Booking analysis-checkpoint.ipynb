Data Analysis and Visualizatiion

````python
# Calculate the percentage distribution of canceled (1) and not canceled (0) bookings.
# value_counts(normalize=True) converts counts into proportions, helping us understand
# the cancellation rate in the dataset.

cancelled_perc = df['is_canceled'].value_counts (normalize = True)
print(cancelled_perc)
````

````python
# Create a bar chart to visualize the number of canceled vs not canceled reservations.
# figsize sets the size of the plot.
# title adds a heading to the chart.
# plt.bar draws bars for both categories using value_counts() from the 'is_canceled' column.
# edgecolor='k' adds a black border to the bars for better visibility.
# width=0.7 controls the bar thickness.
plt.figure(figsize=(5,4))
plt.title('Reservation status Counts')
plt.bar(['Not Cancelled','Cancelled'], df['is_canceled'].value_counts(), edgecolor='k', width=0.7,
 color=['green', 'red'])                # different colors
````

````python
plt.figure(figsize=(8,4))

ax1 = sns.countplot(
    x='hotel',
    hue='is_canceled',
    data=df,
    palette='Blues'       # correct spelling
)

# Fix legend position
ax1.legend(bbox_to_anchor=(1,1))

plt.title('Reservation Status in Different Hotels', size=20)
plt.xlabel('Hotel')
plt.ylabel('Number of Reservations')

plt.show()
````


## **1️⃣ `plt.figure(figsize=(8,4))`**

### ✔ Meaning:

This creates a new empty plot with a size of **8 inches wide and 4 inches tall**.

### ✔ Why:

So the chart looks large, clean, and readable.
Otherwise, the default small size may look cramped.

---

## **2️⃣ `ax1 = sns.countplot(...)`**

This line draws a **countplot**, which counts how many rows fall into each category.

### ✔ Meaning of each parameter:

### **`x='hotel'`**

We are grouping the data by **hotel types**:

* City Hotel
* Resort Hotel

So bars will appear for each hotel type.

---

### **`hue='is_canceled'`**

This separates each bar into two colors:

* `0` = Not canceled
* `1` = Canceled

### ✔ Why:

This shows cancellation patterns within each hotel type.
For example:

* City Hotel: many cancellations
* Resort Hotel: fewer cancellations

Very useful in EDA.

---

### **`data=df`**

We are telling Seaborn to use our dataset `df`.

---

### **`palette='Blues'`**

This sets the color style for the bars.

✔ Blue shades
✔ Light = One category
✔ Dark = Another category

---

# 3️⃣ `ax1.legend(bbox_to_anchor=(1,1))`

### ✔ Meaning:

Moves the legend (Canceled / Not canceled) **outside** the plot, to the right.

Without this:
Legend overlaps the bars → chart looks messy.

---

# 4️⃣ `plt.title(...), plt.xlabel(...), plt.ylabel(...)`

### ✔ Meaning:

Adds proper titles and axis labels.

* Title = what the chart is about
* xlabel = what categories appear on x-axis
* ylabel = what numbers the bars represent

### ✔ Why:

Makes visualization professional and easy to understand.

---

# 5️⃣ `plt.show()`

Displays the final chart.

---

# 📌 SUMMARY

| Code                | Meaning                                   | Why we do it                        |
| ------------------- | ----------------------------------------- | ----------------------------------- |
| `plt.figure()`      | Create a new plot                         | Make it big and clean               |
| `sns.countplot()`   | Draw bars grouped by hotel & cancellation | Compare hotel cancellation behavior |
| `hue='is_canceled'` | Split each bar into canceled/not canceled | Understand patterns                 |
| `palette='Blues'`   | Use blue colors                           | Make it visually clear              |
| `legend(...)`       | Move legend                               | Avoid overlapping                   |
| Labels & title      | Add information                           | Make chart readable                 |
| `plt.show()`        | Show plot                                 | Display output                      |


````python
#selects only the rows where hotel = 'Resort Hotel'.
#counts how many bookings were canceled vs. not canceled, but in percentage (%) form. normalize=True → shows proportion instead of raw count

resort_hotel =df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize =True)
````

````python
city_hotel =df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize =True)
````

````python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
````

````python
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate in City and resort', size=30)

plt.plot(resort_hotel.index, resort_hotel['adr'], label='Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label='City Hotel')
plt.legend(fontsize=20)
plt.show()

````

````python
df['month'] = df['reservation_status_date'].dt.month

plt.figure(figsize=(16,8))
ax1 = sns.countplot(
    x='month',
    hue='is_canceled',
    data=df,
    palette='bright'       
)

ax1.legend(['Not Cancelled','Cancelled'], bbox_to_anchor=(1,1))

plt.title('Reservation Status per Month', size=20)
plt.xlabel('Month')
plt.ylabel('Number of Reservations')

plt.show()

````



## **1️⃣ `df['month'] = df['reservation_status_date'].dt.month`**

### ✔ Meaning:

Extracts the **month number (1–12)** from the reservation date column.

Example:

* 2017-01-15 → month = 1
* 2017-08-04 → month = 8

### ✔ Why:

To analyze **seasonal trends** in bookings and cancellations.

Hotels care about:

* Peak months
* Low months
* Cancellation spikes (rainy season, festivals, holidays)

---

## **2️⃣ `sns.countplot(...)`**

Creates a **bar plot** that shows:

* Months on X-axis
* Number of bookings on Y-axis
* Different colors showing **Canceled vs Not Canceled**

### ✔ Why:

To find patterns like:

* Which month has highest bookings?
* Which months have high cancellation rates?
* Seasonal demand behavior

This is extremely useful in hotel analytics.

---

## **3️⃣ `ax1.legend(['Not Cancelled','Cancelled'], bbox_to_anchor=(1,1))`**

### ✔ Meaning:

We manually rename categories:

* `0 → Not Cancelled`
* `1 → Cancelled`

And move the legend right side of the graph.

### ✔ Why:

To make the chart readable + professional.

---

````python
plt.figure(figsize=(15,8))
plt.title('ADR per Month', fontsize=30)

sns.barplot(
    x='month',
    y='adr',
    data=df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index(),
    hue='month',
    palette='bright',
    legend=False
)

plt.show()
````

````python


# Filter canceled and not canceled
canceled_data = df[df['is_canceled'] == 1]
not_canceled_data = df[df['is_canceled'] == 0]

# Group by date and calculate mean ADR
canceled_df_adr = canceled_data.groupby('reservation_status_date')['adr'].mean()
not_canceled_df_adr = not_canceled_data.groupby('reservation_status_date')['adr'].mean()

# Plotting
plt.figure(figsize=(20,6))
plt.plot(not_canceled_df_adr.index, not_canceled_df_adr.values, label='Not Canceled', color='blue')
plt.plot(canceled_df_adr.index, canceled_df_adr.values, label='Canceled', color='orange')

plt.title('Average Daily Rate')
plt.xlabel('Reservation Date')
plt.ylabel('ADR')
plt.legend()
plt.grid(True)

# Format x-axis for better date display
plt.gcf().autofmt_xdate()  # rotates dates automatically

plt.show()
````

````python
# Define date range
start_date = pd.to_datetime('2016-01-01')
end_date = pd.to_datetime('2017-12-31')

# Canceled data
canceled_data = df[df['is_canceled'] == 1]
canceled_df_adr = canceled_data.groupby('reservation_status_date')['adr'].mean()
canceled_df_adr = canceled_df_adr.reset_index()  # Convert to DataFrame
canceled_df_adr = canceled_df_adr.sort_values('reservation_status_date')
canceled_df_adr = canceled_df_adr[(canceled_df_adr['reservation_status_date'] >= start_date) & 
                                  (canceled_df_adr['reservation_status_date'] <= end_date)]

# Not canceled data
not_canceled_df_adr = df[df['is_canceled'] == 0].groupby('reservation_status_date')['adr'].mean()
not_canceled_df_adr = not_canceled_df_adr.reset_index()  # Convert to DataFrame
not_canceled_df_adr = not_canceled_df_adr.sort_values('reservation_status_date')
not_canceled_df_adr = not_canceled_df_adr[(not_canceled_df_adr['reservation_status_date'] >= start_date) & 
                                          (not_canceled_df_adr['reservation_status_date'] <= end_date)]

# Plotting with correct x and y (dates vs ADR)
plt.figure(figsize=(20,6))
plt.title('Average Daily Rate (2016-2017)')
plt.plot(not_canceled_df_adr['reservation_status_date'], not_canceled_df_adr['adr'], label='Not Canceled')
plt.plot(canceled_df_adr['reservation_status_date'], canceled_df_adr['adr'], label='Canceled')
plt.xlabel('Reservation Status Date')
plt.ylabel('Average Daily Rate (ADR)')
plt.xticks(rotation=45)  # Rotate dates to prevent overlap
plt.legend()
plt.tight_layout()
plt.show()
````
